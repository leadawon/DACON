```python
import pandas as pd

# CSV 파일 경로들의 리스트
csv_files = []

csv_files1 = [
    "ft_efficientnet_v2_m_submit_seed_18.csv",
    "ft_efficientnet_v2_m_submit_seed_4.csv",
    "ft_efficientnet_v2_m_submit_seed_11.csv",
    "ft_efficientnet_v2_m_submit_seed_6.csv",
    "ft_efficientnet_v2_m_submit_seed_0.csv",
    
    
    
]





csv_files2 = [
    "ft_efficientnet_v2_m_submit_seed_64.csv",
    "ft_efficientnet_v2_m_submit_seed_66.csv",
    "ft_efficientnet_v2_m_submit_seed_50.csv",
    
]



for i in csv_files1:
    csv_files.append(r"./seed/"+i)
for i in csv_files2:
    csv_files.append(r"./seed4/"+i)


#csv_files = csv_files1 + csv_files2

print(csv_files)


# CSV 파일들을 DataFrame으로 로드하여 리스트에 저장
dfs = [pd.read_csv(file) for file in csv_files]

# 파일의 개수의 절반 계산
half_count = len(csv_files) / 2

# 첫 번째 DataFrame을 기준으로 다른 DataFrame들의 label을 더함
total_labels = dfs[0]['label'].copy()
for df in dfs[1:]:
    total_labels += df['label']

# 더한 label 값이 파일 개수의 절반보다 크거나 같으면 1, 아니면 0으로 설정
final_labels = (total_labels >= half_count).astype(int)

# 최종 DataFrame 생성
final_df = dfs[0].copy()
final_df['label'] = final_labels

# 결과를 새로운 CSV 파일로 저장
final_df.to_csv('./ensemble/final_result1.csv', index=False)
```
